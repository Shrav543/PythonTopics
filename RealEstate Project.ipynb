```python
def isFloat(val):
    try:
        float(val)
    except:
        return False
    return True
```

```python
df3[~df3["total_sqft"].apply(isFloat)].head()   # ~ will return the opposite of the result
```

```python
def takeaverageofrange(x1):
    try:
        if '-' in x1:
            a, b = tuple(map(float, (x1.split('-'))))
            return (a+b) / 2
        else:
            return float(x1)
    except:
        return None
```

```python
df4 = df3.copy()
df4["total_sqft"]= df4["total_sqft"].apply(takeaverageofrange)
```
